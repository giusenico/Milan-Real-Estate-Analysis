# Real Estate Price Analysis in Milan

## Objective
In this section, we will calculate the price distribution indices (mean, median, and variance) for both the macrozones and microzones of Milan. These indices will help us understand the overall pricing landscape and highlight any significant differences between the areas.

```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
data = pd.read_csv('dataset.csv')

data = data.drop('Unnamed: 0', axis=1)

# Displaying the first few rows to verify the data structure
data.head()
```

## Calculating Price Distribution Indices for Macrozones

In this section, we will calculate the mean, median, and variance of prices for each macrozone in Milan.

```python
# Calculating the mean, median, and variance of prices by macrozone
price_stats_macrozone = data.groupby('macrozone')['price'].agg(['mean', 'median', 'var'])

# Displaying the results
price_stats_macrozone
```

## Calculating Price Distribution Indices for Microzones

Next, we will calculate the mean, median, and variance of prices for each microzone in Milan.

```python
# Calculating the mean, median, and variance of prices by microzone
price_stats_microzone = data.groupby('microzone')['price'].agg(['mean', 'median', 'var'])

# Displaying the results
price_stats_microzone
```

## Exploring Zones with Notable Price Disparities

In this section, we will explore macrozones and microzones that show notable differences between the mean and median prices, or that have particularly high variance. This analysis will help identify the presence of outliers or understand the dynamics of price variability in different areas.

```python
# Identify macrozones with a mean significantly higher than the median
macrozone_disparity = price_stats_macrozone[price_stats_macrozone['mean'] > 1.5 * price_stats_macrozone['median']]
print("Macrozones with significant disparity between mean and median:")
print(macrozone_disparity)

# Identify macrozones with high variance
macrozone_high_variance = price_stats_macrozone[price_stats_macrozone['var'] > price_stats_macrozone['var'].mean()]
print("\nMacrozones with high variance in prices:")
print(macrozone_high_variance)

# Identify microzones with a mean significantly higher than the median
microzone_disparity = price_stats_microzone[price_stats_microzone['mean'] > 1.5 * price_stats_microzone['median']]
print("\nMicrozones with significant disparity between mean and median:")
print(microzone_disparity)

# Identify microzones with high variance
microzone_high_variance = price_stats_microzone[price_stats_microzone['var'] > price_stats_microzone['var'].mean()]
print("\nMicrozones with high variance in prices:")
print(microzone_high_variance)
```

## Visualizing Price Disparities and Variance

In this section, we will create visualizations to better understand the price distributions in the identified macrozones and microzones. We will use boxplots and histograms to highlight disparities, outliers, and general variability in property prices.

```python
import matplotlib.pyplot as plt

# Filtering data to include only macrozones with high variance
high_variance_macrozone_names = macrozone_high_variance.index
filtered_data_macrozone = data[data['macrozone'].isin(high_variance_macrozone_names)]

# Creating boxplot for high variance macrozones
plt.figure(figsize=(12, 8))
filtered_data_macrozone.boxplot(column='price', by='macrozone', vert=False)
plt.title('Price Distribution for Macrozones with High Variance')
plt.xlabel('Price (€)')
plt.ylabel('Macrozone')
plt.suptitle('')  # Remove the default title to clean up the plot
plt.show()
```

```python
# Filtering data for macrozone with significant disparity between mean and median
disparity_macrozone_name = macrozone_disparity.index
filtered_data_disparity = data[data['macrozone'].isin(disparity_macrozone_name)]

# Creating histogram for macrozone with significant disparity
plt.figure(figsize=(10, 6))
plt.hist(filtered_data_disparity['price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Price Distribution for Macrozone with Significant Disparity')
plt.xlabel('Price (€)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Filtering data to include only microzones with high variance
high_variance_microzone_names = microzone_high_variance.index
filtered_data_microzone = data[data['microzone'].isin(high_variance_microzone_names)]

# Creating boxplot for high variance microzones
plt.figure(figsize=(12, 8))
filtered_data_microzone.boxplot(column='price', by='microzone', vert=False)
plt.title('Price Distribution for Microzones with High Variance')
plt.xlabel('Price (€)')
plt.ylabel('Microzone')
plt.suptitle('')  # Remove the default title to clean up the plot
plt.show()
```

## Handling Outliers

In this section, we will identify and remove outliers from the dataset. The goal is to perform a correlation analysis on a dataset that represents the general market without the influence of extreme values. We will then create a separate dataset to analyze the properties that are significantly more expensive (the outliers).

```python
import numpy as np

# Calculating the threshold for outliers based on the z-score
# We'll consider properties with a z-score greater than 3 as outliers

# Adding a new column with the z-score of the price
data['price_zscore'] = (data['price'] - data['price'].mean()) / data['price'].std()

# Filtering data to exclude outliers (considering |z-score| <= 3)
data_no_outliers = data[np.abs(data['price_zscore']) <= 3]

# Creating a dataset with only the outliers for future analysis
data_outliers = data[np.abs(data['price_zscore']) > 3]

# Dropping the 'price_zscore' column from both datasets, as it's not needed anymore
data_no_outliers = data_no_outliers.drop(columns=['price_zscore'])
data_outliers = data_outliers.drop(columns=['price_zscore'])

# Displaying the number of properties in each dataset
print(f"Number of properties without outliers: {len(data_no_outliers)}")
print(f"Number of outliers: {len(data_outliers)}")
```

## Milan Home Price Index (Milan HPI) - Based on Price Segments

The **Milan Home Price Index (HPI)** is a measure designed to track the overall trend of home prices in Milan. By calculating the HPI for different price segments, we gain a better understanding of how prices fluctuate across different areas of the city. We divide the properties into **three price categories**: low, middle, and high.

### Purpose:
- The HPI helps to **monitor trends** in the housing market in Milan.
- By segmenting the index by price bands, we can assess how property values evolve across different market segments (e.g., affordable housing, mid-range properties, luxury properties).

### How the HPI is Calculated:
We calculate the HPI for three price segments using **base values** derived from external data on Milan’s real estate market in 2023.

### Price Segments and Base Values:
- **Low Price Segment** (outer areas like Bisceglie, Baggio, Olmi): Average price of **€2.833/m²** (October 2023).
- **Middle Price Segment** (intermediate areas like Città Studi, Susa): Average price of **€5.528/m²** (October 2023).
- **High Price Segment** (central areas like Centro, Garibaldi-Moscova-Porta Nuova): Average price of **€10.600/m²** (October 2023).

The sources for these values are based on the latest real estate reports from Immobiliare.it and Fanpage.it&#8203;:contentReference[oaicite:0]{index=0}.

### Formula:
The formula for calculating the Milan HPI is:

\begin{equation}
HPI = \left( \frac{P_{\text{current}}}{P_{\text{base}}} \right) \times 100
\end{equation}

Where:
- **Pcurrent**: Median price across properties in a given price segment.
- **Pbase**: Base value for each price segment, as specified above.

### Sources:
- Immobiliare.it, [Osservatorio Immobiliare](https://www.immobiliare.it) - Report ottobre 2023.
- Fanpage.it, [Quanto costa comprare una casa a Milano](https://www.fanpage.it/milano/)&#8203;:contentReference[oaicite:1]{index=1}.


```python
# Step 1: Calculate the average price for each macrozone
macrozone_prices_mean = data.groupby('macrozone')['price'].mean()

# Step 2: Define a function to assign each macrozone to a price segment based on the average price
def assign_price_segment(price):
    if price < 2833:
        return 'Low'
    elif 2833 <= price < 5528:
        return 'Middle'
    else:
        return 'High'

# Step 3: Apply the function to macrozones to assign the price segment
macrozone_prices_mean_segmented = macrozone_prices_mean.apply(assign_price_segment)

# Step 4: Map the results back into the original dataframe
data['price_segment'] = data['macrozone'].map(macrozone_prices_mean_segmented)

# Step 5: Define price segments and corresponding base values
price_segments = {
    'Low': 2833,  # Low price segment
    'Middle': 5528,  # Middle price segment
    'High': 10600  # High price segment
}

# Step 6: Calculate the base value for each macrozone
data['base_value'] = data['price_segment'].map(price_segments)

# Step 7: Calculate the HPI for each macrozone using the corresponding base value
data['HPI'] = (data['price'] / data['base_value']) * 100

# Step 8: Group by macrozone and calculate the average HPI for each macrozone
hpi_macrozone_segmented = data.groupby('macrozone')['HPI'].mean()

# Display the results
print(hpi_macrozone_segmented)
```

```python
# Step 1: Calculate the quartiles of the HPI to dynamically define the segments
q1 = hpi_macrozone_segmented.quantile(0.25)  # 25th percentile
q2 = hpi_macrozone_segmented.quantile(0.50)  # 50th percentile (median)
q3 = hpi_macrozone_segmented.quantile(0.75)  # 75th percentile

# Step 2: Group the macrozones based on the HPI quartiles
low_hpi = hpi_macrozone_segmented[hpi_macrozone_segmented <= q1]  # Low segment (<= Q1)
middle_hpi = hpi_macrozone_segmented[(hpi_macrozone_segmented > q1) & (hpi_macrozone_segmented <= q3)]  # Middle segment (Q1 < HPI <= Q3)
high_hpi = hpi_macrozone_segmented[hpi_macrozone_segmented > q3]  # High segment (> Q3)

# Step 3: Calculate the average HPI for each segment
low_hpi_mean = low_hpi.mean()
middle_hpi_mean = middle_hpi.mean()
high_hpi_mean = high_hpi.mean()

# Step 4: Display the results
print(f"Average HPI for Low Price Segment (HPI <= {q1:.2f}): {low_hpi_mean:.2f}")
print(f"Average HPI for Middle Price Segment ({q1:.2f} < HPI <= {q3:.2f}): {middle_hpi_mean:.2f}")
print(f"Average HPI for High Price Segment (HPI > {q3:.2f}): {high_hpi_mean:.2f}")
```

```python
# Step 1: Define the labels for the bands.
price_segments = ['Low Price Segment', 'Middle Price Segment', 'High Price Segment']
hpi_means = [low_hpi_mean, middle_hpi_mean, high_hpi_mean]

# Step 2: Create the bar graph.
plt.figure(figsize=(10, 6))
plt.bar(price_segments, hpi_means, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average HPI by Price Segment')
plt.ylabel('Average HPI')
plt.xlabel('Price Segment')
plt.show()
```

```python
# Step 1: Sort the macrozones according to HPI.
hpi_macrozone_sorted = hpi_macrozone_segmented.sort_values()

# Step 2: Create the bar graph.
plt.figure(figsize=(12, 8))
hpi_macrozone_sorted.plot(kind='bar', color='lightblue')
plt.title('HPI by Macrozona')
plt.ylabel('HPI')
plt.xlabel('Macrozona')
plt.xticks(rotation=90)
plt.show()
```

## Milan Housing Affordability Index (HAI) by Macrozone

The **Milan Housing Affordability Index (HAI)** measures whether an average household in Milan can afford a median-priced property. The HAI for each macrozone is calculated based on the median property prices we have collected in our dataset.

### Formula:

\begin{equation}
HAI = \left( \frac{\text{Median Household Income}}{\text{Required Income to Purchase a House}} \right) \times 100
\end{equation}

Where:
- **Required Income** is calculated as:
\begin{equation}
\frac{\text{Median Home Price} \times \text{Mortgage Interest Rate}}{12 \times \text{Mortgage Multiplier}}
\end{equation}

### Data Used:
1. **Median Property Prices by Macrozone**: Median prices for each macrozone are sourced from our dataset.
2. **Median Household Income in Milan**: **€30,000** per year, according to the Bank of Italy&#8203;:contentReference[oaicite:0]{index=0}.
3. **Average Mortgage Interest Rate**: The current average mortgage interest rate in Italy for 20-year loans is **3.73%**, according to Numbeo&#8203;:contentReference[oaicite:1]{index=1}.
4. **Mortgage Multiplier**: **25%** of household income is typically allocated for mortgage payments.

### Interpretation:
- If **HAI > 100**, it means the average household can afford to purchase a median-priced home.
- If **HAI < 100**, it means the median-priced home is unaffordable for the average household.

### Sources:
- Property prices: Data from our project dataset.
- Median Household Income: Bank of Italy, [Survey on Household Income and Wealth](https://www.bancaditalia.it)&#8203;:contentReference[oaicite:2]{index=2}.
- Mortgage Interest Rate: [Numbeo - Mortgage Interest Rates in Italy](https://www.numbeo.com)&#8203;:contentReference[oaicite:3]{index=3}.

```python
# General data
median_income = 30000  # Median annual household income (in euros)
interest_rate = 0.0373  # Average mortgage interest rate (3.73%)
mortgage_multiplier = 0.25  # Portion of income allocated to mortgage (25%)

hai_results = []

# Iterate over the median prices for each macrozone
for macrozone, median_price in median_price_macrozone.items():
    # Calculate the required income for each macrozone
    monthly_payment = (median_price * interest_rate) / 12  # Monthly mortgage payment
    required_income = monthly_payment / mortgage_multiplier  # Required income for the mortgage

    # Calculate the HAI for the macrozone
    hai = (median_income / required_income) * 100

    # Aggiungere i risultati alla lista
    hai_results.append({'Macrozone': macrozone, 'HAI': hai})
    
    # Display results for each macrozone
    print(f"Macrozone: {macrozone}, Milan HAI: {hai:.2f}")

hai_macrozone_df = pd.DataFrame(hai_results)
```

```python
# Create a bar chart for HAI by macrozone
macrozone = list(median_price_macrozone.keys())  # List of macrozones
hai_values = []  # List to store HAI values

# Iterate over the median prices for each macrozone and calculate HAI
for macrozone_name, median_price in median_price_macrozone.items():
    monthly_payment = (median_price * interest_rate) / 12  # Monthly mortgage payment
    required_income = monthly_payment / mortgage_multiplier  # Required income for the mortgage
    hai = (median_income / required_income) * 100  # Calculate HAI
    hai_values.append(hai)  # Add HAI to the list

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(macrozone, hai_values, color='skyblue')
plt.xlabel('Macrozone')
plt.ylabel('HAI Value')
plt.title('Housing Affordability Index (HAI) by Macrozone')
plt.xticks(rotation=90)
plt.show()
```

```python
# Define the price ranges
low_price_zone = {k: v for k, v in median_price_macrozone.items() if v <= 300000}
mid_price_zone = {k: v for k, v in median_price_macrozone.items() if 300000 < v <= 600000}
high_price_zone = {k: v for k, v in median_price_macrozone.items() if v > 600000}

# Calculate the HAI for each price range
hai_low = sum([(median_income / ((v * interest_rate) / 12 / mortgage_multiplier)) * 100 for v in low_price_zone.values()]) / len(low_price_zone)
hai_mid = sum([(median_income / ((v * interest_rate) / 12 / mortgage_multiplier)) * 100 for v in mid_price_zone.values()]) / len(mid_price_zone)
hai_high = sum([(median_income / ((v * interest_rate) / 12 / mortgage_multiplier)) * 100 for v in high_price_zone.values()]) / len(high_price_zone)

# Create a bar chart for the price ranges
plt.figure(figsize=(8, 5))
plt.bar(['Low Price', 'Mid Price', 'High Price'], [hai_low, hai_mid, hai_high], color=['lightgreen', 'yellow', 'salmon'])
plt.xlabel('Price Range')
plt.ylabel('Average HAI Value')
plt.title('Housing Affordability Index (HAI) by Price Range')
plt.show()
```
